Дана функция f(x) = 5x^2 + 10x - 30

1. Определить корни                                     +

2. Найти интервалы, на которых функция возрастает       +

3. Найти интервалы, на которых функция убывает          +

4. Построить график                                     +

5. Вычислить вершину                                    +

6. Определить промежутки, на котором f > 0              +

7. Определить промежутки, на котором f < 0              +


```python
import math
from sympy import *
from sympy.plotting import plot as plt
# from sympy import Symbol
init_printing()

x = Symbol('x')
fx = -5 * x ** 2 + 10 * x - 150
fx
```

**Построение графика**

```python
p = plt(fx, (x,-100,100))
```

**Вершина**

```python
xh = solve(diff(fx))[0]
yh = -5 * xh ** 2 + 10 * xh - 150
[xh, yh]
```

Определить промежутки, на котором f > 0

```python
if solve(fx>0) == False:
    print ("График функции не имеет положительных значений ")
else: 
    print(solve(fx>0))
```

Определить промежутки, на котором f < 0

```python
if solve(fx<0) == False:
    print ("График функции не имеет отрицательных значений ")
else: 
    print(solve(fx<0))
```

Определить корни

```python
x1, x2 = solve(fx)
print('первый корень', round(x1.evalf(), 2),'\nвторой корень', round(x2.evalf(), 2))
```

2. Найти интервалы, на которых функция возрастает

```python
solve(diff(fx)>0)
```

3. Найти интервалы, на которых функция убывает 

```python
solve(diff(fx)<0)
```
